```python
#!/usr/bin/env python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bayes(mp,start,actions):
    j = start[0]
    k = start[1]
    bel = np.zeros((10,20))
    mp[j,k] = 8
    bel[j,k] = 1
    belx = 1
    path = [(j,k)]
    print("Each step taken is shown by 8")
    print("Starting point")
    print(mp)
    for i in actions:
        if i=='R':
            if (k==19):
                bel[j,k] = 1
            if (mp[j,k+2]==1):
                bel[j,k] = 0.2*belx
                bel[j,k+1] = 1 - bel[j,k]
                bel[j,k+2] = 0
            if (mp[j,k+1]==1):
                bel[j,k] = 1
                bel[j,k+1] = 0
                bel[j,k+2] = 0
            else:
                bel[j,k] = 0.2*belx
                bel[j,k+1] = 0.6*belx
                bel[j,k+2] = 0.2*belx
            belx = max(bel[j,k],bel[j,k+1],bel[j,k+2])
            j,k = np.where(bel==belx)
        if i=='L':
            if (k==0):
                bel[j,k] = 1
            if (mp[j,k-2]==1):
                bel[j,k] = 0.2*belx
                bel[j,k-1] = 1 - bel[j,k]
                bel[j,k-2] = 0
            if (mp[j,k-1]==1):
                bel[j,k] = 1
                bel[j,k-1] = 0
                bel[j,k-2] = 0
            else:
                bel[j,k] = 0.2*belx
                bel[j,k-1] = 0.6*belx
                bel[j,k-2] = 0.2*belx
            belx = max(bel[j,k],bel[j,k-1],bel[j,k-2])
            j,k = np.where(bel==belx)
        if i=='U':
            if (j==0):
                bel[j,k] = 1
            if (mp[j-2,k]==1):
                bel[j,k] = 0.2*belx
                bel[j-1,k] = 1 - bel[j,k]
                bel[j-2,k] = 0
            if (mp[j-1,k]==1):
                bel[j,k] = 1
                bel[j-1,k] = 0
                bel[j-2,k] = 0
            else:
                bel[j,k] = 0.2*belx
                bel[j-1,k] = 0.6*belx
                bel[j-2,k] = 0.2*belx
            belx = max(bel[j,k],bel[j-1,k],bel[j-2,k])
            j,k = np.where(bel==belx)
        if i=='D':
            if (j==9):
                bel[j,k] = 1
            if (mp[j+2,k]==1):
                bel[j,k] = 0.2*belx
                bel[j+1,k] = 1 - bel[j,k]
                bel[j+2,k] = 0
            if (mp[j+1,k]==1):
                bel[j,k] = 1
                bel[j+1,k] = 0
                bel[j+2,k] = 0
            else:
                bel[j,k] = 0.2*belx
                bel[j+1,k] = 0.6*belx
                bel[j+2,k] = 0.2*belx
            belx = max(bel[j,k],bel[j+1,k],bel[j+2,k])
            j,k = np.where(bel==belx)
        mp[j,k] = 8
        path.append((j,k))
        print("Action: ", i)
        print(mp)
    print(path)
    return bel
```

```python
mp = np.zeros((10,20))
start = (8,5)
start = (9 - start[1],start[0]) #inverted according to the matrix
bel = bayes(mp,start,"RURLUD")
```

```python
plt.matshow(bel)
plt.colorbar()
plt.show()
```
